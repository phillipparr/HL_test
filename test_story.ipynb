```python
import pandas as pd
from collections import Counter
import numpy as np
funding = pd.read_csv('crunchbase_funding.csv')
funding
```

```python
def get_counts(df, column_name):
    column = df[column_name]
    counts = Counter(column)
    return counts.most_common()

def get_count_list_columns(df, column_name):
    list_ = df[column_name]
    new_list = []
    for i in list_:
        clean = i.lstrip('[').rstrip(']').split(', ')  
        for i in clean:
            new_list.append(i.strip('\''))
    return Counter(new_list).most_common()
```

```python
# Cities
cities = get_counts(funding, 'location_city')
cities
```

```python
beijing = funding.loc[lambda funding: funding['location_city'] == 'Beijing']
sf = funding.loc[lambda funding: funding['location_city'] == 'San Francisco']
nyc = funding.loc[lambda funding: funding['location_city'] == 'New York']
london = funding.loc[lambda funding: funding['location_city'] == 'London']
cambridge = funding.loc[lambda funding: funding['location_city'] == 'Cambridge']
```

```python
sf_industry_counts = get_count_list_columns(sf, 'fundedorg_categories')
sf_industry_counts
```

```python
# Types of funding
funding_type = get_counts(funding, 'funding_type')
funding_type
```

```python
# Compare money raised
venture = funding.loc[lambda funding: funding['funding_type'] == 'venture']
venture_sum = np.nansum(venture['money_raised_usd'])
print("Venture:", '${:,.2f}'.format(venture_sum))
seed = funding.loc[lambda funding: funding['funding_type'] == 'seed']
seed_sum = np.nansum(seed['money_raised_usd'])
print("Seed:", '${:,.2f}'.format(seed_sum))
grant = funding.loc[lambda funding: funding['funding_type'] == 'grant']
grant_sum = np.nansum(grant['money_raised_usd'])
print("Grant:", '${:,.2f}'.format(grant_sum))
angel = funding.loc[lambda funding: funding['funding_type'] == 'angel']
angel_sum = np.nansum(angel['money_raised_usd'])
print("Angel:", '${:,.2f}'.format(angel_sum))
product_crowdfunding = funding.loc[lambda funding: funding['funding_type'] == 'product_crowdfunding']
product_crowdfunding_sum = np.nansum(product_crowdfunding['money_raised_usd'])
print("Product Crowdfunding:", '${:,.2f}'.format(product_crowdfunding_sum))
```
